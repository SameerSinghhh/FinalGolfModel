```python
#Imports
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


import xgboost as xgb
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

```

```python
#Read in cleaned data (Already scaled)
df = pd.read_excel('G.xlsx', 'RidgeTrain')
df = df.set_index('Unnamed: 0')

df.head(5)
```

```python
# 70% Train / 30% Test
trainnum = 2397

df_trained = df[:trainnum]
df_test = df[trainnum:]
df_test.head(5)
```

```python
#SPLIT
x = df_trained.iloc[:trainnum, :-6] 
y = df_trained['Top_5']
    
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
# XGBOOST
xgb_model = xgb.XGBClassifier(objective='binary:logistic', random_state=42, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
xgb_predictions = xgb_model.predict_proba(X_test)[:, 1]  
xgb_score = xgb_model.score(X_test, y_test)

# SVM
svm_model = SVC(probability=True, random_state=42, kernel='rbf')
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict_proba(X_test)[:, 1]  
svm_score = svm_model.score(X_test, y_test)

# LOGISTIC REGRESSION
logistic_model = LogisticRegression(random_state=42, penalty='l2', max_iter=1000)
logistic_model.fit(X_train, y_train)
logistic_predictions = logistic_model.predict_proba(X_test)[:, 1]  
logistic_score = logistic_model.score(X_test, y_test)

# RANDOM FOREST
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict_proba(X_test)[:, 1]
rf_score = rf_model.score(X_test, y_test)

# GRADIENT BOOSTING
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)
gb_predictions = gb_model.predict_proba(X_test)[:, 1]
gb_score = gb_model.score(X_test, y_test)

# Print all model accuracies
print("XGBoost Model Accuracy:", xgb_score)
print("SVM Model Accuracy:", svm_score)
print("Logistic Regression Model Accuracy:", logistic_score)
print("Random Forest Model Accuracy:", rf_score)
print("Gradient Boosting Model Accuracy:", gb_score)
```

```python
k_folds = 5

# XGBOOST
xgb_model = xgb.XGBClassifier(objective='binary:logistic', random_state=42, eval_metric='logloss')
xgb_cv_scores = cross_val_score(xgb_model, X_train, y_train, cv=k_folds, scoring='accuracy')

# SVM
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardizing the data
    ('svm', SVC(probability=True, random_state=42, kernel='rbf'))
])
svm_cv_scores = cross_val_score(svm_pipeline, X_train, y_train, cv=k_folds, scoring='accuracy')

# LOGISTIC REGRESSION
logistic_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardizing the data
    ('logistic', LogisticRegression(random_state=42, penalty='l2', max_iter=1000))
])
logistic_cv_scores = cross_val_score(logistic_pipeline, X_train, y_train, cv=k_folds, scoring='accuracy')

# RANDOM FOREST
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_cv_scores = cross_val_score(rf_model, X_train, y_train, cv=k_folds, scoring='accuracy')

# GRADIENT BOOSTING
gb_model = GradientBoostingClassifier(random_state=42)
gb_cv_scores = cross_val_score(gb_model, X_train, y_train, cv=k_folds, scoring='accuracy')

# Print all model cross-validated accuracies
print("XGBoost Model CV Mean Accuracy:", xgb_cv_scores.mean())
print("SVM Model CV Mean Accuracy:", svm_cv_scores.mean())
print("Logistic Regression Model CV Mean Accuracy:", logistic_cv_scores.mean())
print("Random Forest Model CV Mean Accuracy:", rf_cv_scores.mean())
print("Gradient Boosting Model CV Mean Accuracy:", gb_cv_scores.mean())
```

```python
scaler = MinMaxScaler()

x = df_trained.iloc[:trainnum, :-6]
y = df_trained['Top_20']
    
    
# Define the parameter grid for Random Forest
parameters = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2']  # Number of features to consider when looking for the best split
}

# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid = GridSearchCV(estimator=rf_model, param_grid=parameters, cv=5, scoring='accuracy')

# Fit the model using GridSearchCV
grid.fit(X_train, y_train)

# Print the best score and the best parameters
print("Best Accuracy:", grid.score(X_test, y_test))
print("Best Parameters:", grid.best_params_)
```

```python
# Define the best parameters as found from GridSearchCV
best_params = {
    'n_estimators': 200,
    'max_depth': None,
    'max_features': 'auto',
    'min_samples_split': 10,
    'min_samples_leaf': 2
}

# Initialize the Random Forest model with the best parameters
rf_model = RandomForestClassifier(
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    max_features=best_params['max_features'],
    min_samples_split=best_params['min_samples_split'],
    min_samples_leaf=best_params['min_samples_leaf'],
    random_state=42
)

# Fit the model to the training data
model = rf_model.fit(x, y)
```

```python
model_predictions = list(model.predict_proba(df_test.iloc[:, :14]))

df_outright = pd.DataFrame(list(zip([y * 100 for x,y in model_predictions], df_test.index))).sort_values(0, ascending = False).set_index(1)
df_outright.to_csv('OutrightWinner.csv')
```
